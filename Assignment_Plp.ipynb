```python
import pandas as pd
import numpy as np

# Creating a mock dataset
np.random.seed(42)
species = np.random.choice(['setosa', 'versicolor', 'virginica'], 150)
sepal_length = np.random.normal(loc=5.8, scale=0.8, size=150)
sepal_width = np.random.normal(loc=3.0, scale=0.4, size=150)
petal_length = np.random.normal(loc=3.8, scale=1.2, size=150)
petal_width = np.random.normal(loc=1.2, scale=0.6, size=150)

data = {
    'species': species,
    'sepal_length': sepal_length,
    'sepal_width': sepal_width,
    'petal_length': petal_length,
    'petal_width': petal_width
}

df = pd.DataFrame(data)

# Show the first few rows
print(df.head())
```

```python
# Basic statistics of the numerical columns
print(df.describe())

# Group by species and get mean of each numerical column
print(df.groupby('species').mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Line chart (simulating a time trend)
df['index'] = range(len(df))
sns.lineplot(x='index', y='sepal_length', data=df)
plt.title("Sepal Length Trend (Fake Time Series)")
plt.show()

# Bar chart: Average Petal Length by Species
sns.barplot(x='species', y='petal_length', data=df)
plt.title("Average Petal Length by Species")
plt.show()

# Histogram: Sepal Length Distribution
sns.histplot(df['sepal_length'], bins=10, kde=True)
plt.title("Sepal Length Distribution")
plt.show()

# Scatter Plot: Sepal vs. Petal Length
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal Length vs. Petal Length")
plt.show()
```
